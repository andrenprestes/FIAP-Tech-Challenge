```python
import json
import pandas as pd
```

```python

# Caminho para o JSON
caminho_arquivo = "/content/drive/MyDrive/Colab Notebooks/FIAP_POS_ML/FASE5/dados/applicants.json"

# Carrega o JSON inteiro (só faça isso se tiver certeza que cabe na memória)
with open(caminho_arquivo, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Transforma o dicionário de candidatos em uma lista de registros
registros = []
for id_candidato, info in data.items():
    registro = {"id": id_candidato}
    for secao, campos in info.items():
        if isinstance(campos, dict):
            for chave, valor in campos.items():
                registro[f"{secao}__{chave}"] = valor
        else:
            registro[secao] = campos
    registros.append(registro)

# Cria o DataFrame
df = pd.DataFrame(registros)

# Mostra as primeiras linhas
print(df.head())
```

```python
ids_dataframe = df["id"].astype(str)  # IDs no DataFrame como string
ids_json = set(data.keys())           # IDs do JSON original

ids_df_set = set(ids_dataframe)

# Comparação
faltando_no_df = ids_json - ids_df_set
extras_no_df = ids_df_set - ids_json

print("IDs no JSON:", len(ids_json))
print("IDs únicos no DataFrame:", len(ids_df_set))
print("IDs faltando no DataFrame:", len(faltando_no_df))
print("IDs extras no DataFrame:", len(extras_no_df))
```
